```python
import re
import pandas as pd
from math import log
import matplotlib.pyplot as plt

def parse_ising_data(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
    data = []
    for line in lines:
                      #Ising: E=0.2864, m=-0.076;  Heat bath: E=0.3831, T = 2.099; E_tot=0.3210.023
        m = re.match(r'Ising: E=([\d\.\-]+), m=(\-?[\d\.]+);  Heat bath: E=([\d\.\-]+), T = ([\d\.\-]+); E_tot=([\d\.\-]+).*', line.strip())
        if m:
            data.append({
                'Ising_E': float(m.group(1)),
                'Ising_m': float(m.group(2)),
                'HeatBath_E': float(m.group(3)),
                'HeatBath_T': 1/log((1/float(m.group(3))) - 1),
                'E_tot': float(m.group(5))
            })
    return pd.DataFrame(data)

def plot_ising(df, x_axis, y_axis, l, lab=None):
    plt.plot(df[x_axis], df[y_axis], marker='x',label=lab)

def show_plot_ising(x_axis, y_axis, l):
    plt.xlim(l)
    plt.xlabel(x_axis)
    plt.ylabel(y_axis)
    plt.title(f'{y_axis} vs {x_axis}')
    plt.grid(True)
    plt.legend()
    plt.show()


# Example usage:

plt.figure(figsize=(20,7))

rt=[]
rm=[]
with open('pp.txt','r') as fh:
    for line in fh.readlines():
        a,b=tuple(map(float,line.strip().split(' ')))
        rt.append(a)
        rm.append(b)

plt.plot(rt, rm, marker='o',label='real')

df = parse_ising_data('Isingdata30.txt')
plot_ising(df, 'HeatBath_E', 'Ising_m', (0.1,0.5),'30x100')
df = parse_ising_data('Isingdata50.txt')
plot_ising(df, 'HeatBath_E', 'Ising_m', (0.1,0.5),'50x100')
df = parse_ising_data('Isingdata70.txt')
plot_ising(df, 'HeatBath_E', 'Ising_m', (0.1,0.5),'70x100')
df = parse_ising_data('Isingdata90.txt')
plot_ising(df, 'HeatBath_E', 'Ising_m', (0.1,0.5),'90x100')

show_plot_ising('HeatBath_E', 'Ising_m', (0.1,0.5))



#print(list(zip(df['HeatBath_T'],map(lambda a:-a,df['Ising_m'])))[::10])
```
